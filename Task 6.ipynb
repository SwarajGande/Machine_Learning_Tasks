# Assignment for visualization using Python Matplotlib Library

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Days = [1,2,3,4,5,6,7,8]
Speed = [60,62,61,58,56,57,46,63]
plt.plot(Days,Speed,"r--")
plt.xlabel("Days")
plt.ylabel("Speed")
plt.title("Car Speed Measurement")
plt.show()
```

```python
plt.plot(Days,Speed,"g--",marker = "o",ms= 15,linewidth =3,mec = "black",mfc = "orange")
plt.xlabel("Days")
plt.ylabel("Speed")
plt.title("Car Speed Measurement")
plt.show()
```

```python
days=[1,2,3,4,5,6,7,8]
max_speed=[80,91,92,88,77,79,76,75]
min_speed=[42,43,40,42,33,36,34,35]
avg_speed=[46,58,57,56,40,42,41,36]
```

```python
plt.subplot(3,1,1)
plt.plot(days,max_speed,"g--",marker = "o",label ="Max_speed")
plt.xlabel("Days")
plt.ylabel("max_speed")
plt.title("Car Speed Measurement")
plt.legend()
plt.grid()
plt.show()
plt.subplot(3,1,2)
plt.plot(days,min_speed,"r--",marker = "o",label ="Min_speed")
plt.xlabel("Days")
plt.ylabel("min_speed")
plt.title("Car Speed Measurement")
plt.legend()
plt.grid()
plt.show()
plt.subplot(3,1,3)
plt.plot(days,avg_speed,"b--",marker = "o",label ="Avg_speed")
plt.xlabel("Days")
plt.ylabel("avg_speed")
plt.title("Car Speed Measurement")
plt.legend()
plt.grid()
plt.show()
```

```python
t = np.linspace(0,5*np.pi,100)
x1 = np.sin(t)
x2 = np.cos(t)
plt.plot(t,x1,"b--",linewidth = 1.5,label = "sin")
plt.plot(t,x2,"r-",linewidth = 1.5,label = "cos")
plt.xlabel("Time")
plt.ylabel("Magnitude")
plt.legend(loc = "best",shadow = 1)
plt.show()
```

```python
Languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
Popularity = [56, 39, 34, 34, 29]
Security = [44 ,36 ,55, 50, 42]
```

```python
plt.bar(Languages,Popularity,width = 0.5)
plt.show()
```

```python
plt.bar(Languages,Security,width = 0.5)
plt.show()
```

```python
x = np.arange(len(Languages))
plt.bar(x-0.2,Popularity,label = "Popularity",width = 0.4)
plt.bar(x+0.2,Security,label = "Security",width = 0.4)
plt.xticks(x,Languages)
plt.xlabel("Languages")
plt.ylabel("Popularity & Security")
plt.title("Popularity and Security of major programming Languages")
plt.legend(loc = "best",shadow = 1)
plt.show()
```

```python
plt.barh(x-0.2,Popularity,0.4,label = "Popularity")
plt.barh(x+0.2,Security,0.4,label = "Security")
plt.yticks(x,Languages)
plt.ylabel("Languages")
plt.xlabel("Popularity & Security")
plt.title("Popularity and Security of major programming Languages")
plt.legend(loc = "best",shadow = 1)
plt.show()
```

```python
Marks = np.array([61,86,42,46,73,95,65,78,53,92,55,69,70,49,72,86,64])
Below_Average = Marks[np.logical_and(Marks >= 40,Marks < 60)]
Average = Marks[np.logical_and(Marks >= 60,Marks < 80)]
Excellent= Marks[np.logical_and(Marks >= 80,Marks < 100)]
```

```python
plt.hist(Marks,color = 'orange',edgecolor = "black")
plt.xlabel("Distribution")
plt.ylabel("Marks")
plt.title("Student Marks")
plt.vlines([40,60,80,100],ymin = 0,ymax = 4,linestyle = "dashed")
plt.show()
print("Below Average Students :",Below_Average.size)
print("Average Students :",Average.size)
print("Excellent Students :",Excellent.size)
```
